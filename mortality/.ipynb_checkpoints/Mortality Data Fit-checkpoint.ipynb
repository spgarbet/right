# Mortality Data Distribution Fitting

## Load data from http://mortality.org (2013)

```r
lt.male = read.csv("mltper_1x1.txt", sep="", skip=2,
                   header=TRUE, as.is=TRUE)
lt.female = read.csv("fltper_1x1.txt", sep="", skip=2,
                     header=TRUE, as.is=TRUE)

# Narrow to 2013
yy = 2013
lt.male = subset(lt.male,Year==yy)
lt.female = subset(lt.female,Year==yy)

# Cleanup
lt.male[lt.male$Age=="110+",]$Age = 110
lt.female[lt.female$Age=="110+",]$Age = 110
lt.male$Age = as.numeric(lt.male$Age)
lt.female$Age = as.numeric(lt.female$Age)

# Derive CDF
lt.male$cdf = with(lt.male,1-lx/100000)
lt.female$cdf = with(lt.female,1-lx/100000)

lt = rbind.data.frame(
      cbind.data.frame(gender="female",lt.female),
      cbind.data.frame(gender="male",lt.male))

# Create a simulator from raw cdf

probDeath <- function(A=0, gender="male")
{
    Ages <- 1:110
    cdf = lt[lt$gender==gender,]$cdf
    prDeath = diff(cdf)
    
    prDeath[Ages>A]/sum(prDeath[Ages>A]) # Reormalize
}

simulate <- function(A=0, gender="male", NN=10000)
{
    Ages <- 1:110
    prDeath.c <- probDeath(A, gender)
    if (length(prDeath.c)>1)
        ss = sample(Ages[Ages>A],NN,prob=prDeath.c,replace=TRUE)
    if (length(prDeath.c)==1)
        ss = Ages[Ages>A]

    ss
}
```

## Weibull Fit Vs. Simulated from CDF Death from Age 0

```r
ss <- simulate()
hist(ss, freq=FALSE, breaks=20, main="Weibull vs. Simulated Males Age[0]")
curve(dweibull(x, 5.537834, 82.60117), add=TRUE, col='red')
ks.test(ss, rweibull(10000, 5.537834, 82.60117))
```

The Weibull fit of the data is sub optimal. Basically, there are very few deaths in the Weibull before 20, even really before 25.
Then the spike post 80 is completely cut off.  

The infant mortality presents a problem. Ideally, only data above 1 would be usable with a single parameterized distribution. Simulations involving infant mortality will require more advanced methods, than are proposed here for a single distribution fit. The rest of this will focus on Mortality from age 40, which is used in the simulation.

```r
ss <- simulate(40)

h<-hist(ss, freq=FALSE, breaks=20, main="Weibull vs. Simulated Males Age[40]")
curve(dweibull(x-40, 3.519033, 44.68335), add=TRUE, col='red')
pfit <- function(x) {pweibull(x-40, 3.519033, 44.68335)}
ks.test(ss, "pfit")

plot(40:110, c(cumsum(probDeath(40)),1), type='s',
     main="Census versus Weibull Fit", xlab="Age", ylab="CDF of Mortality")
curve(pfit, add=TRUE, col='red')

library(entropy)

KL.plugin(h$counts, round(diff(pfit(h$breaks))*sum(h$counts)))
```

Likewise again, using the Weibull distribution a patient who enters a simulation at age 40, is very unlikely to die via the Weibull curve before around 45. However, the histogram shows mass present and patients should be dieing during this period of the simulation.

## Try a Gompertz Distribution

```r
library(flexsurv)
library(survival)

# Have to expand data into single discrete events for gompertz fitting
df2 <- lt.male[41:111, ]
df3 <- data.frame(Age=rep(df2$Age, df2$dx)-40)
df3$Death <- 1

s1 <- with(df3, Surv(Age, Death))
f1 <- flexsurvreg(s1 ~ 1, dist="gompertz")
f1
```

```r
h<-hist(ss, freq=FALSE, main="Simulation vs. Gompertz", xlab="Male Deaths After 40")
curve(dgompertz(x-40, 8.96e-2, 1.74e-3) , add=TRUE, col='red')
pfit <- function(x) {pgompertz(x-40, 8.96e-2, 1.74e-3)}
ks.test(ss, "pfit")
plot(40:110, c(cumsum(probDeath(40)),1), type='s',
     main="Census versus Gompertz Fit", xlab="Age", ylab="CDF of Mortality")
curve(pfit(x), add=TRUE, col='red')
```

In addition a Generalized Beta Type 2 (4-parameter) fit was tried, and it was approximately as good as the Weibull.
